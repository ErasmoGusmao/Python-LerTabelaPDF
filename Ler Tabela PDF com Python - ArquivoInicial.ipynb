### Mas nem sempre a tabela vai vir bonitinha

```python
lista_tabelas2 = tabula.read_pdf("ed_result_obj_mjsp.pdf", pages ="all")
print(len(lista_tabelas2))
```

```python
tabela = lista_tabelas2[0]
display(tabela)
```

Tratando a formatação da taela

```python
tabela = lista_tabelas2[0]
tabela.columns = tabela.iloc[0]

# Divido a última coluna em duas separadas por um espaço (" ")
tabela[[0,1]] = tabela["Variação percentual"].str.split(" ", expand = True)

# Returando a linha 0 e as linhas pós a liha 8
tabela = tabela[1:9]

# Ajustado a primeira coluna
tabela=tabela.set_index("R$ milhões")

# Ajustando o cabeçalho
tabela.columns = tabela.iloc[0]
tabela = tabela[1:]
tabela= tabela.drop("1T21/4T20 1T21/1T20", axis = 1)
display(tabela)
```
